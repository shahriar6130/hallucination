```python
from pathlib import Path
import pandas as pd

# ----------------------------
# Load Dataset
# ----------------------------
DATA_PATH = Path("../data/sample.csv")

df = pd.read_csv(DATA_PATH)

print("=" * 60)
print("DATASET OVERVIEW")
print("=" * 60)

print(f"Rows: {len(df)}")
print(f"Columns: {len(df.columns)}")

print("\nColumns:")
print(df.columns.tolist())

print("\n")

# ----------------------------
# Missing Values
# ----------------------------
print("=" * 60)
print("MISSING VALUES")
print("=" * 60)

print(df.isnull().sum())

print("\n")

# ----------------------------
# Label Distribution
# ----------------------------
print("=" * 60)
print("LABEL DISTRIBUTION")
print("=" * 60)

# Check if 'label' column exists before counting values
if "label" in df.columns:
    print(df["label"].value_counts())
    print()
    print(df["label"].value_counts(normalize=True))
else:
    print("⚠️ 'label' column not found in this dataset.")

print("\n")

# ----------------------------
# NULL Context
# ----------------------------
print("=" * 60)
print("NULL CONTEXT")
print("=" * 60)

# Safety check for 'context' column
if "context" in df.columns:
    null_context = (
        df["context"]
        .fillna("")
        .astype(str)
        .str.upper()
        .eq("NULL")
        .sum()
    )
    print("Rows with context == NULL:", null_context)
else:
    print("⚠️ 'context' column not found. Skipping text evaluation for 'context'.")

print("\n")

# ----------------------------
# Duplicate Rows
# ----------------------------
print("=" * 60)
print("DUPLICATES")
print("=" * 60)

print(df.duplicated().sum())

print("\n")

# ----------------------------
# Text Length
# ----------------------------
print("=" * 60)
print("TEXT LENGTH")
print("=" * 60)

target_cols = ["context", "prompt_bn", "response_bn"]

for col in target_cols:
    # Only calculate length if the column actually exists
    if col in df.columns:
        lengths = df[col].fillna("").astype(str).str.len()
        print(f"\n📊 Length Distribution for: '{col}'")
        print(lengths.describe())
    else:
        print(f"\n⏩ Skipping '{col}' (Column not found in CSV)")

print("\n")

# ----------------------------
# Preview
# ----------------------------
print("=" * 60)
print("FIRST FIVE ROWS")
print("=" * 60)

print(df.head())
```
